```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection(name="my_flower_collection")
```

```python
collection.add(ids=["id1", "id2", "id3"],documents=["A vibrant red rose","A sunny yellow tulip","A pure white lily"])
```

```python
results = collection.get()
print(results)
```

```python
results = collection.query(
    query_texts=["yellow flowers"], # Chroma will embed this for you
    n_results=2 # how many results to return
)
print(results)
```

```python
results = collection.query(
    query_texts=["pink flowers"], # Chroma will embed this for you
    n_results=1 # how many results to return
)
print(results)
```
